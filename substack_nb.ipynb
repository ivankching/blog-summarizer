```python
from substack_api import Post, Newsletter, User
from datetime import datetime, timezone
from pathlib import Path
import shutil
```

```python
directory = Path('post_content')

for item in directory.iterdir():
    if item.is_file() or item.is_symlink():
        item.unlink()
    elif item.is_dir():
        shutil.rmtree(item)
```
